```python
import pandas as pd
df = pd.read_csv('../data/subset_protein_mapped_new.csv')
columns_to_drop = [
    'domain_sequence1',
    'domain_start2', 'domain_end2', 'domain_sequence2',
    'domain_start3', 'domain_end3', 'domain_sequence3',
    'domain_start4', 'domain_end4', 'domain_sequence4',
    'domain_start5', 'domain_end5', 'domain_sequence5',
    'domain_start6', 'domain_end6', 'domain_sequence6',
    'cath_domain_start2', 'cath_domain_end2',
    'cath_domain_start3', 'cath_domain_end3',
    'cath_domain_start4', 'cath_domain_end4',
    'cath_domain_start5', 'cath_domain_end5',
    'cath_domain_start6', 'cath_domain_end6'
]

# Drop the columns
# axis=1 or axis='columns' indicates that you are dropping columns
# inplace=False (default) returns a new DataFrame without modifying the original
df = df.drop(columns=columns_to_drop, axis=1)
df
```

```python
target_label_col_list = ["class", "architecture", "topology", "homology"]
df['cath'] = df[target_label_col_list[0]].astype(str)
for col in target_label_col_list[1:]:
    df['cath'] = df['cath'] + '.' + df[col].astype(str)
```

```python
import pandas as pd
import numpy as np

# Simulate a DataFrame with 'cath' column (as in the previous interaction)
np.random.seed(42) # for reproducibility
# Get the value counts of the 'cath' column
cath_value_counts = df["cath"].value_counts()

# Filter the value counts to show only entries where the count is 20
counts_equal_to_20 = cath_value_counts[cath_value_counts == 15]

df_new = df[df["cath"].isin( counts_equal_to_20.index.to_list())]
```

```python
df_new["domain_start"] = df_new["domain_start"].astype("Int64")

# Convert 'domain_end' column
df_new["domain_end"] = df_new["domain_end"].astype("Int64")
```
