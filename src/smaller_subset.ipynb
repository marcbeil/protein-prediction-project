```python
import pandas as pd
df = pd.read_csv('../data/subset_protein_mapped_enhanced.csv')
df
```

```python
df[df['protein_length'] < 600].to_csv("../data/subset_protein_mapped_enhanced_limited_len_600.csv", index=False)
```

```python
target_label_col_list = ["class", "architecture", "topology", "homology"]
df['cath'] = df[target_label_col_list[0]].astype(str)
for col in target_label_col_list[1:]:
    df['cath'] = df['cath'] + '.' + df[col].astype(str)

df["protein_length"] = df["protein_sequence"].str.len()
```

```python
df["protein_id"] = df["domain_id"].str[:4]
```

```python
df["domain_start"] = df["domain_start"].astype("int")
df["domain_end"] = df["domain_end"].astype("int")
```

```python
df
```

```python
df.to_csv("../data/subset_protein_mapped_enhanced", index=False)
```
